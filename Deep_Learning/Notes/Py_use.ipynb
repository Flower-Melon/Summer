## 2.先从排序开始，自己重新编写一遍
* 选择排序和冒泡排序

````python
# 选择排序实现
b = [12,16,17,19,304,2025]
for i in range((len(b))):
    max_index = i
    for j in range(i+1,len(b)):
        if b[j] > b[max_index]:
            max_index = j
    b[i],b[max_index] = b[max_index],b[i]
print(b)

# 冒泡排序实现
c = [12,16,17,19,304,2025]
for i in range(len(c)):
    for j in range(len(c)-1,i-1,-1):
        if c[j] < c[j-1]:
            c[j-1],c[j] = c[j],c[j-1]
print(c)
        
````

* 快速排序

````python
import random

# 快速排序实现
def position(lists,left,right):
    base = lists[left]
    while left < right:
        while left < right and lists[right] >= base:
            right -= 1
        lists[left] = lists[right]
        while left < right and lists[left] <= base:
            left += 1
        lists[right] = lists[left]
    lists[left] = base
    return left

def quick_sort(lists,left,right):
    if left < right:
        pos = position(lists,left,right)
        quick_sort(lists,left,pos-1)
        quick_sort(lists,pos+1,right)

# 生成随机列表
random.seed(10)
lists = [random.randint(0,1000) for i in range(10)]
quick_sort(lists,0,len(lists)-1)
print(lists)
````

## 5.实战一下
* 完成一个简单的线性回归编写

````python
import numpy as np
import random
import matplotlib.pyplot as plt
````

* 生成数据集

````python
def generate_data(w,b,num):
    X = np.random.random((num,len(w)))
    y = X @ w + b
    y += np.random.normal(0,0.01,size=y.shape) #加入噪声
    return X,y.reshape(-1,1)

# 真实值
true_w = np.array([2,-3.4])
true_b = 4.2

features,label = generate_data(true_w,true_b,1000)
plt.scatter(features[:,1],label)
````

* 随机读取数据集
> 这里使用yield定义了一个生成器get_data

````python
def get_data(batch_size,features,label):
    num = len(features)
    indexs = np.arange(num) # 索引
    # random.seed(100)
    random.shuffle(indexs)# 随机打乱
    
    for i in range(0,num,batch_size):
        batch_index = np.array(indexs[i:min(i+batch_size,num)])
        yield features[batch_index],label[batch_index]

batch_size = 10
````

* 初始化模型参数

````python
w = np.random.normal(0,0.01,size=(2,1))
b = np.zeros(1)
````

* 定义模型

````python
def linear(X,w,b):
    return X @ w + b
````

* 训练

````python
lr = 0.01
epoch = 100

for i in range(epoch):
    for X,y in get_data(batch_size,features,label):
        l = linear(X,w,b) - y # 计算误差 
        
        # 求解梯度
        grad_w = 2 * (X.T @ l) / batch_size 
        grad_b = 2 * l.sum() / batch_size
        # 更新参数
        w -= lr * grad_w
        b -= lr * grad_b
        
        # 也可以使用增广的方法
        # X_b = np.c_[np.ones((X.shape[0], 1)), X]  # 在X的第一列添加1
        # grad = 2 * X_b.T @ l / batch_size  # 求解梯度
        # w -= lr * grad[1:]  # 更新参数（跳过第一行）
        # b -= lr * grad[0]  # 更新参数


print(f'w的误差为：{w.reshape(true_w.shape) - true_w}')
print(f'b的误差为：{b - true_b}')
````
